```python
#!/bin/python3

import math
import os
import random
import re
import sys
import csv
import collections
from collections import defaultdict


#
# Complete the 'arrangingRules' function below.

# The function is expected to return a STRING_ARRAY.
# The function accepts STRING_ARRAY rules as parameter.
#
rules = []
Values = defaultdict(list)
Kval = {}
arr = []
Results = []
def arrangingRules(rules):
    with open('census.csv', newline='', encoding="utf8") as csvfile:
        csv_file = csv.reader(csvfile, delimiter=',')
        line = 0
        for row in csv_file:
            for column in row:
                key, val = column.partition("=")[::2]
                Values[line].append([key,val])
                Kval[key] = val
            line += 1
        #print(line)
        Rs = rules
        ctr = 0
        for row in Rs:
            X, Y = row.partition("=>")[::2]

            X = X.replace("{","")
            X = X.replace("}","")
            Y = Y.replace("{","")
            Y = Y.replace("}","")

            Xs = X.split (",")
            Ys = Y.split(",")

            X_keys = []
            X_vals = []
            Y_keys = []
            Y_vals = []
            for i in Xs:
                X, Y = i.partition("=")[::2]
                X_keys.append(X)
                X_vals.append(Y)

            for i in Ys:
                X, Y = i.partition("=")[::2]
                Y_keys.append(X)
                Y_vals.append(Y)

            
            X_count = 0
            Y_count = 0
            for i in range(0,30162):
                cntx = 0
                for j in range(0,len(X_vals)):
                    s = [X_keys[j],X_vals[j]]
                    if s in Values[i]:
                        cntx += 1
                cnty = 0
                for j in range(0,len(Y_vals)):
                    s = [Y_keys[j],Y_vals[j]]
                    if s in Values[i]:
                        cnty += 1
                if cntx == len(X_vals):
                    X_count += 1
                    if cnty == len(Y_vals):
                        Y_count += 1

            fval = float(Y_count)/float(X_count)
            pair = (fval,ctr)
            arr.append(pair)
            ctr += 1
        
        #arr.sort(key=lambda x:float(10),reverse=True)
        for i in range(0,len(arr)-1):
            for j in range(0,len(arr)-i-1):
                if arr[j] < arr[j+1]:
                    temp = arr[j];
                    arr[j] = arr[j+1]
                    arr[j+1] = temp
        i = 0
        for a,b in arr:
            Results.append(rules[b])
            i += 1
        print(Results)
        print(arr)
    return Results
    

if __name__ == '__main__':
   # fptr = open(os.environ['OUTPUT_PATH'], 'w')

    rules_count = int(input().strip())

    rules = []

    for _ in range(rules_count):
        rules_item = input()
        rules.append(rules_item)

    result = arrangingRules(rules)

    #fptr.write('\n'.join(result))
    #fptr.write('\n')

#    fptr.close()
```
