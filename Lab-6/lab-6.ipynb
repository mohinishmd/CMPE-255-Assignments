```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

X = [ 17.5 , 22, 29.5, 44.5, 64.5, 80]
Y = [ 38 , 36, 24 , 20, 18, 28]

plt.xlabel("Age", fontsize=18)
plt.ylabel("Deaths", rotation=90, fontsize=18)
plt.plot(X, Y, "b.")
plt.axis([0, 90, 5, 50])
plt.show()
```

```python
X_b = np.c_[np.ones((6, 1)), X]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)

X_new = np.array([[0], [80]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.xlabel("Age", fontsize=18)
plt.ylabel("Deaths", rotation=90, fontsize=18)
plt.plot(X, Y, "b.")
plt.axis([0, 85, 5, 50])
plt.show()
```

```python
X_new = np.array([[40]])
X_new_b = np.c_[np.ones((1, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
print("Age = 40, Deaths = ",y_predict)

X_new = np.array([[60]])
X_new_b = np.c_[np.ones((1, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
print("Age = 60, Deaths = ",y_predict)
```
