```python
def transform_to_features(data):
    from sklearn.feature_extraction.text import CountVectorizer
    vectorizer = CountVectorizer(
        analyzer = 'word',
        lowercase = False,
    )
    features = vectorizer.fit_transform(
        data
    )
    features_nd = features.toarray()
    return features_nd


def train_then_build_model(data_labels, features_nd, comments, labels):
    from sklearn.model_selection import train_test_split
    # TODO : set training % to 80%.
    X_train, X_test, y_train, y_test  = train_test_split(
        features_nd, 
        data_labels,
        train_size=0.80, 
        random_state=0)
    
    X_data_train, X_data_test, y_data_train, y_data_test  = train_test_split(
        comments, 
        labels,
        train_size=0.80, 
        random_state=0)
    
    from sklearn.linear_model import LogisticRegression
    from sklearn.model_selection import cross_val_score
    from sklearn import datasets, linear_model
    from sklearn import cross_validation

    log_model = LogisticRegression()

    log_model = log_model.fit(X=X_train, y=y_train)
    y_pred = log_model.predict(X_test)

    scores = cross_validation.cross_val_score(log_model, features_nd, labels, cv=5)
    lines = 0
    ctr = 0
    print("Same Output: Positive reviews")
    for row in X_data_test:
        if y_data_test[lines] == '5' and y_pred[lines] == 'pos':
            print(row)
            ctr += 1
        lines += 1
        if ctr >= 3:
            break;

    print()
    print()
    ctr=0
    lines = 0
    print("Same Output: Negative reviews")
    for row in X_data_test:
        if y_data_test[lines] == '1' and y_pred[lines] == 'neg':
            print(row)
            ctr += 1
        lines += 1
        if ctr >= 3:
            break;
            
    print()
    print()
    lines = 0
    ctr = 0
    print("Opposite Output: 5 stars, Negative predictions")
    for row in X_data_test:
        if y_data_test[lines] == '5' and y_pred[lines] == 'neg':
            print(row)
            ctr += 1
        lines += 1
        if ctr >= 3:
            break;

    print()
    print()
    ctr=0
    lines = 0
    print("Opposite Output: 1 Star, Positive predictions")
    for row in X_data_test:
        if y_data_test[lines] == '1' and y_pred[lines] == 'pos':
            print(row)
            ctr += 1
        lines += 1
        if ctr >= 3:
            break;
            
            
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import precision_score
    from sklearn.metrics import recall_score
    
    print()
    print("Accuracy=",accuracy_score(y_test, y_pred))
    print("Precision=",precision_score(y_test, y_pred,average='weighted'))
    print("Recall=",recall_score(y_test, y_pred,average='weighted'))
    precision = precision_score(y_test, y_pred,average='weighted')
    recall = recall_score(y_test, y_pred,average='weighted')
    F1 = 2 * (precision * recall) / (precision + recall)
    print("F1 score=", F1)
    print("Cross validation score (cv iterator = 5)= ",scores)

    


features_nd = transform_to_features(comments)
X_data_test = []
train_then_build_model(labels, features_nd , comments, stars)
X = pd.DataFrame(comments)
Y = pd.Series(labels)



```
